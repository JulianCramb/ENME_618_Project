```python
import laspy
import numpy as np
from utilities import load_lidar, visualize, simulate_noise, print_lidar
```

# Noise Simulation for Urban LiDAR

```python
pc1 = 'Data/segmented_points_urban.las'
#pc1 = 'Data/300-5050_2015.las'

xyz, intensity, classification, scan_angle, header = load_lidar(pc1)

print(xyz.shape)
```

```python
visualize(xyz, classification)
```

```python
xyz_noisy, classification_noisy, intensity_noisy = simulate_noise(
    xyz,
    intensity=intensity,
    classification=classification,
    scan_angle=None,
    num_per_type=0,
    num_outliers = 50000,
    apply={
        "range": True,
        "horizontal": True,
        "scan_angle": True,
        "surface": True,
        "outliers": True,
        "intensity": True,
    }
)

print(xyz_noisy.shape)
```

```python
visualize(xyz_noisy, classification_noisy)
```

```python
print_lidar(xyz_noisy, intensity_noisy, classification_noisy, 'Data/300-5050_2015_w_Noise.las', header)
```
